## Analysis by Borough

```python
boroPop = pd.read_csv("Population_by_Borough_NYC.csv", index_col="Borough")

for i in range(1,6):
    val = boroPop["2010"].iloc[i]
    val = val.replace(',' , '')
    boroPop["2010"].iloc[i] = val
    
print(boroPop["2010"].iloc[1:6].astype(np.int64))
```

```python
boroPop
```

```python
boroPop.drop("NYC Total", axis=0, inplace=True)
```

```python
np.array(boroPop["2010"].iloc[1:6])#.astype(np.int64)
```

```python
# "# crimes" / "borough population" * 1000 = crimes per thousand
boroPop["boro1000"] = np.array(crimeData["BORO_NM"].value_counts())/np.array(boroPop["2010"]).astype(np.int64)
boroPop["boro1000"] = boroPop["boro1000"] * 1000
```

```python
# Crimes per thousand in each borough
boroPop["boro1000"]
```

# Plot crimes coordinates on NYC map 

```python
ax_map = sns.relplot(x="X_COORD_CD", y="Y_COORD_CD", hue="LAW_CAT_CD", data=crimeData.iloc[0:1000]);
```

```python
with sns.axes_style("white"):
    sns.jointplot(x="X_COORD_CD", y="Y_COORD_CD", data=crimeData.iloc[0:1000], kind="hex", color="k");
```

```python
cmap = sns.palplot(sns.light_palette((210, 90, 60), input="husl"));
sns.jointplot(x="X_COORD_CD", y="Y_COORD_CD", data=crimeData.iloc[0:1000], kind="kde", cmap=cmap);
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.jointplot(x="X_COORD_CD", y="Y_COORD_CD", data=crimeData.iloc[0:1000], kind="kde", n_levels=40, shade=True, cmap=cmap);
```
